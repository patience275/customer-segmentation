## preprocessing

```python
df=df.drop(columns=['Customer ID'])
```

```python
df.columns
```

```python
cat_cols=[['Gender', 'City', 'Membership Type', 'Discount Applied','Satisfaction Level']]
```

```python
from sklearn.preprocessing import OneHotEncoder
genderencoder=OneHotEncoder()
encoded_data=genderencoder.fit_transform(df[['Gender']])
gender_encoded_df=pd.DataFrame(encoded_data.toarray(),columns=genderencoder.get_feature_names_out(['Gender']))
df=pd.concat([df.drop(columns=['Gender']),gender_encoded_df],axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
cityencoder=OneHotEncoder()
encoded_data=cityencoder.fit_transform(df[['City']])
city_encoded_df=pd.DataFrame(encoded_data.toarray(),columns=cityencoder.get_feature_names_out(['City']))
df=pd.concat([df.drop(columns=['City']),city_encoded_df],axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
membershipencoder=OneHotEncoder()
encoded_data=membershipencoder.fit_transform(df[['Membership Type']])
membership_encoded_df=pd.DataFrame(encoded_data.toarray(),columns=membershipencoder.get_feature_names_out(['Membership Type']))
df=pd.concat([df.drop(columns=['Membership Type']),membership_encoded_df],axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
discountencoder=OneHotEncoder()
encoded_data=discountencoder.fit_transform(df[['Discount Applied']])
discount_encoded_df=pd.DataFrame(encoded_data.toarray(),columns=discountencoder.get_feature_names_out(['Discount Applied']))
df=pd.concat([df.drop(columns=['Discount Applied']),discount_encoded_df],axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
satisfactionencoder=OneHotEncoder()
satisfaction_data=satisfactionencoder.fit_transform(df[['Satisfaction Level']])
satisfaction_encoded_df=pd.DataFrame(satisfaction_data.toarray(),columns=satisfactionencoder.get_feature_names_out(['Satisfaction Level']))
df=pd.concat([df.drop(columns=['Satisfaction Level']),satisfaction_encoded_df],axis=1)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df=pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
```

```python
df.head()
```

## model training

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

K=range(1,11)
inertia=[]
for k in K:
  kmeans=KMeans(n_clusters=k,random_state=42, n_init=10)
  kmeans.fit(df)
  inertia.append(kmeans.inertia_)

plt.plot(K, inertia, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal K')
plt.grid(True)
plt.show()
```

```python
kmeans=KMeans(n_clusters=6,random_state=42)
kmeans.fit(df)
labels=kmeans.labels_
df1['clusters']=labels
```

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(df,labels)
print('silhoutte score:',score)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced_data = pca.fit_transform(df)

plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='plasma')
plt.title("Clusters Visualized with PCA")
plt.grid(True)
plt.show()
```
